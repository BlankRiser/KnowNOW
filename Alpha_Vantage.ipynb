````python
import os
import io
import datetime
import requests as req
import numpy as np
import pandas as pd
````

````python
from dotenv import load_dotenv
load_dotenv()
ALPHAVANTAGE_KEY = os.getenv('ALPHAVANTAGE_API_KEY')

# For trading stocks
SYMBOL='IBM' 

# For Cryptocurrencies
MARKET='INR' # Refer physical currency list csv for more
SYMBOL_CRYPTO='1ST' # Refer digital currency list csv for more
````

### use when data is csv

```python

```


````python
# Crypto Daily - Candlestcik

DIGITAL_CURRENCY_DAILY = 'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol='+SYMBOL_CRYPTO+'&market='+MARKET+'&apikey='+ALPHAVANTAGE_KEY+'&datatype=csv'

res = req.get(DIGITAL_CURRENCY_DAILY)
res = res.content
rawData = pd.read_csv(io.StringIO(res.decode('utf-8')))


````

### Candlestick chart

I didnt like the way it showed data(no padding at all), so finding other alternatives

```python
import plotly.graph_objects as go

fig_candlestick = go.Figure(data=[go.Candlestick(x=rawData['timestamp'],
                open=rawData['open (INR)'],
                high=rawData['high (INR)'],
                low=rawData['low (INR)'],
                close=rawData['close (INR)'])])

fig_candlestick.update_layout(
    title = f'{SYMBOL_CRYPTO} in {MARKET} ',
    yaxis_title=f'{SYMBOL_CRYPTO}',
    
)

fig_candlestick.show()
```

````python
import plotly.graph_objects as go

fig_scatter = go.Figure([go.Scatter(x=rawData['timestamp'], y=rawData['high (INR)'])])
fig_scatter.show()
````

````python
import plotly.express as px

fig_line = px.line(rawData, x=rawData['timestamp'], y=['high (INR)'], color="Year")

fig_line.update_layout(
    title = f'{SYMBOL_CRYPTO} in {MARKET} ',
    yaxis_title=f'{SYMBOL_CRYPTO} - high (INR)',

)

fig_line.add_hline(y=rawData['high (INR)'].mean(), line_dash="dot",
              annotation_text="Summed Average",
              annotation_position="bottom right")

fig_line.show()
````
